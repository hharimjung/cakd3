```python
# flower.py

import datetime
import time # 마지막 팁에 sleep을 주기 위해 import.
import random # 마지막에 여러가지 팁을 random으로 추출하기 위해 import.


# 하나의 클래스 안에 선택지 함수들을 생성. 
class Flower:
    def __init__(self, flower_Name, flower_Lang,flower_Color,flower_Meaning): # flower 변수 설정.
        self.flower_Name = flower_Name
        self.flower_Lang = flower_Lang
        self.flower_color = flower_Color
        self.flower_meaning = flower_Meaning
        
    def one(self): # 1-(1)자동 추천 <완료>
        now = datetime.datetime.now() # 현재 해당하는 월 가져오기.
        a = now.month
        print('%d월의 추천 꽃은 %s이며,\n꽃말은 %s입니다.' %(a,flower_Name[a],flower_Lang[a])) 
        # flowers[월],[꽃],[꽃말]
        
    def two(self,month_Chioce): # 1-(2) 달별꽃 <완료>
        print('%d월을 선택하셨습니다. %d월의 추천 꽃은 %s입니다.' %(month_Chioce,month_Chioce,flower_Name[month_Chioce]))
        print("-"*50)
        name = input('꽃말이 궁금한 꽃의 이름을 입력해주세요.> ') # 해당 월의 꽃 리스트 출력 -> 꽃 선택 -> 꽃말 추출
        name_Index = flower_Name[month_Chioce].index(name) # index 함수로 필요한 꽃 리스트들 불러오기.
        print('%s의 꽃말은 %s입니다.' %(name, flower_Lang[month_Chioce][name_Index]))
        
    def three(self, flower_Choice): # 1-(3) 직접선택
        for i in range(1,13):
            if flower_Name[i].count(flower_Choice) != 0: # 선택한 꽃이 해당하는 월에 없으면 a에 데이터를 저장하지않음.
                a = flower_Name[i].index(flower_Choice)
                print('[%s]의 꽃말은 [%s]입니다.' %(flower_Choice, flower_Lang[i][a]))
                break
    
    def four(self,color_Choice): # 1-(4) 색상별 정서적의미
        a = flower_Color.index(color_Choice)
        print('%s을(를) 선택하셨습니다. %s 꽃은 %s효과가 있습니다. '%(color_Choice,color_Choice,self.flower_meaning[a]))

             

# 처음 메뉴와 마지막 팁 함수 생성.
def menu():
    choice = int(input('원하시는 항목을 선택해주세요(자동추천:1, 달별꽃:2, 직접선택:3, 색상별 정서적의미:4)> \n'))
    print("-"*50)
    return choice

def tip(flower_Tip,flower_Tip_Text):
    tip_Num = flower_Tip.index(random.choice(flower_Tip)) 
    # flower_Tip의 리스트요소값를 인덱스값으로 불러오기 위해 index 사용.
    print(flower_Tip[tip_Num])
    for i in range(len(flower_Tip_Text[tip_Num])):
        print(flower_Tip_Text[tip_Num][i])
        time.sleep(1.5)
        
        
# 꽃 이름 리스트
flower_Name = [
[], # 인덱스 0번값을 비워둠. 1 = 1월부터 시작하게끔 설정.
[# 1월
        "수선화","스노우드롭","샤프란"],
[# 2월
        "제비꽃", "월계수","은매화"], 
[# 3월
        "수선화","데이지","박하"],
[# 4월        
        "스위트 피","아네모네","나팔꽃"],
[# 5월
        "은방울꽃","민들레","수련"],
[# 6월
        "장미","쟈스민","백리향"],
[# 7월
        "라벤더","해바라기","초롱꽃"],
[# 8월
        "글라디올러스","진달래","금잔화"],
[# 9월
        "물망초","엉겅퀴","로즈마리"],
[# 10월
        "금잔화","국화","매화"],
[# 11월
        "국화","메리골드","크리스마스로즈"],
[# 12월
        "포인세티아","라벤더","쑥국화"]
]

# 꽃말 리스트
flower_Lang = [
[], # 인덱스 0번값을 비워둠. 1 = 1월부터 시작하게끔 설정.
[# 1월
        "매력과 특별함","희망","후회없는 청춘"],
[# 2월
        "지혜와 희망", "명예","사랑의 속삭임"], 
[# 3월
        "가정의 행복,존경,우정","명랑","미덕"],
[# 4월        
        "사랑,젊음,순결","기대","기쁨"],
[# 5월
        "사랑,열정,아름다움,완벽함","신탁","청순한 마음"],
[# 6월
        "사랑","사랑스러움","용기"],
[# 7월
        "정절,기대","애모","감사"],
[# 8월
        "의지의 힘,명예,기억","사랑의 기쁨","이별의 슬픔"],
[# 9월
        "인내,정신적인 사랑,애정","엄격","기억,추억,헌신"],
[# 10월
        "맹렬함,우아함,헌신","사랑","고결한 마음"],
[# 11월
        "연민,우정,기쁨","이별의 슬픔","추억"],
[# 12월
        "용기,성공,특별한 존재","기대","평화"]
]

# 꽃 색깔
flower_Color = ['빨강', '주황', '노랑', '초록', '파랑', '남색', '보라', '흰색']



# 꽃 정서적 의미
flower_Meaning = ['기분이 우울할 때나 마음이 약해졌을 때 감정을 상승시키는 ',
                  '강한 의지를 가질 수 있는 색으로 타인의 말에 유혹되지않고싶을 때 ',
                  '대화를 부드럽게 만드는 색으로 첫 대면할 경우에 ',
                  '감정적인 대화를 편안하게 이끌어가는 ',
                  '상대의 입장에서 생각해 주고 싶을 때 ',
                  '집중력을 높여줘 명상할 때 ',
                  '격조가 높아 보이며 럭셔리함을 주고싶을 때 ',
                  '고민이 있을 때 기분이 좋아지는 ']

# 꽃과 관련된 팁 <마무리>
flower_Tip = ['꽃 색상 조합하는 꿀팁','꽃을 오래 보관하는 꿀팁','꽃 선물 성공하는 꿀팁']

# 꽃과 관련된 팁_내용 <마무리>
flower_Tip_Text = [
    [
        '1. 중심 꽃과 색상을 하나 정한다.',
        '2. 그 꽃과 잘 어울릴만한 꽃들을 옆에 대본다.',
        '3. 다양한 모양의 꽃을 배치한다.',
        '4. 포인트 색상을 주어 꽃을 더 강조해준다.',
        '** 흰색이나 분홍색 꽃으로 조금 융화시키면 더 좋아요! **'
    ],
    [
        '1. 화병에 담을 때 물에 잠기는 잎은 제거해 주세요.',
        '2. 줄기 끝을 비스듬이 잘라주세요.',
        '3. 물에 설탕을 1L에 50g정도 넣어주세요.',
        '4. 과일 옆이나 직사광선을 피해주세요.',
        '5. 밤에는 시원한 곳에 두는 것이 좋아요.',
        '6. 드라이 플라워로 만들어 오랫동안 꽃을 보관하세요.'
    ],
    [
        '1. 받는 분의 취향에 맞는 꽃을 선택하세요.',
        '2. 받는 분이 일하는 곳으로 꽃배달 이벤트를 해보세요.',
        '3. 꽃다발도 좋지만, 때에 따라서 꽃바구니 선물도 좋아요!'
    ]
]



# 인삿말<프로그램 시작> / 보기좋게 공백과 줄을 사용.
print('{0:^40}'.format(chr(0x1F338)))
print()
print('안녕하세요. 향기를 선물하는 꽃 자판기입니다.')
print()
print('{0:^40}'.format(chr(0x1F338)))
print()
print("-"*50)


# 클래스 함수 선언.
menu_Choice = menu()        
obj = Flower(flower_Name,flower_Lang,flower_Color,flower_Meaning) 


# 처음 메뉴 선택지를 선택하면 나오는 멘트 정리.
if menu_Choice == 1:
    print('자동추천을 선택하셨습니다. 계절을 담은 제철꽃을 추천드릴게요.')
    print()
    obj.one()
elif menu_Choice == 2:
    month_Chioce = int(input('달별꽃을 선택하셨습니다. 원하시는 달을 입력해주세요.> '))
    obj.two(month_Chioce)
    
elif menu_Choice == 3:
    for i in range(1, 13):
        print("%d월 :\t" %(i), flower_Name[i])
    print("-"*50)
    flower_Choice = input('직접선택을 선택하셨습니다. 알아보고싶은 꽃 이름을 입력해주세요.> ')
    obj.three(flower_Choice)
    
else: 
    color_Choice = input('색상별 정서적의미를 선택하셨습니다. 원하시는 색상을 입력해주세요.\n[보기] : 빨강, 주황, 노랑, 초록, 파랑, 남색, 보라, 흰색 >')
    obj.four(color_Choice)
    
# 프로그램 종료 전, 꽃에 관련한 팁 출력
print("-"*50)    
tip(flower_Tip,flower_Tip_Text)

# 마무리 멘트<프로그램 종료>
print("-"*50)
print()
print('{0:^30}'.format('"꽃은 사랑의 진정한 언어입니다."'))
print('{0:^40}'.format(chr(0x1F338)))
```
